```python
def add(x,y):
    return x + y
def subtract(x,y):
    return x - y
def multiply(x,y):
    return x * y
def divide(x,y):
    if y==0:
        print("Error:You cant divide a number by zero")
    else:
        return x/y
def get_number(prompt):
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("Invalid number:Please enter a valid value")
def main():
    while True:
        print("\nChoose the operation")
        print("1.Addition +")
        print("2.Multiplication *")
        print("3.Subtraction -")
        print("4.Division /")
        print("5. Exit q")
        choice = input("Enter choice: ").strip()

        if choice == "q":
            print("Goodbyee")
            break

        if choice not in ["+","/","*","-"]:
            print("Invalid choice.Please enter a valid choice:")
            continue
        
        num1 = int(input("Enter the fisrt number: "))
        num2 = int(input("Enter the secod number: "))
        if choice == "+":
            results = add(num1,num2)
        elif choice == "/":
            results = divide(num1,num2)
        elif choice == "*":
            results = multiply(num1,num2)
        elif choice == "-":
            results =subtract(num1,num2)
        
        print("Results: ",results)
if __name__=="__main__":
    main()
    
              
    
```
